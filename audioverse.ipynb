```python
from transformers import AutoProcessor, AutoModel

processor = AutoProcessor.from_pretrained("suno/bark")
model = AutoModel.from_pretrained("suno/bark")

inputs = processor(
    text=["Hello, my name is Suno. And, uh — and I like pizza. [laughs] But I also have other interests such as playing tic tac toe."],
    return_tensors="pt",
)

speech_values = model.generate(**inputs, do_sample=True)
```

```python
from transformers import pipeline
import scipy


synthesiser = pipeline("text-to-speech", "suno/bark")

speech = synthesiser("Hello, my dog is cooler than you!", forward_params={"do_sample": True})

scipy.io.wavfile.write("bark_out.wav", rate=speech["sampling_rate"], data=speech["audio"])
```

```python
from transformers import pipeline
```

```python
suno = pipeline("text-to-speech", model= "suno/bark-small")
text = """
          I would be testing out the text to speech for the first time.
          It seems to be working pretty well, and I would like to see how much better it can work.
          Here is a little love letter to me:
          " Hey Me,
            I love how well you have worked hard this year 2023. And honestly I can wait to see you grow exponentially in the coming year.
            What do you think about love, or have we given up on that"

          """
```

```python
output_voice = suno(text)
```

```python
from IPython.display import Audio
Audio(output_voice["audio"], rate=output_voice["sampling_rate"])
```
